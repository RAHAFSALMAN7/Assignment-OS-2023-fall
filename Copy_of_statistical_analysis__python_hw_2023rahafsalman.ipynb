```python
from google.colab import files
import io
import pandas as pd
import numpy as np
from scipy.stats import t

# Upload the file
uploaded = files.upload()

# Get the file name
file_name = next(iter(uploaded))

# Read the CSV file into a DataFrame
example_data = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Function 1: calculate_trip_duration
def calculate_trip_duration(data_frame):
    data_frame['pickup_datetime'] = pd.to_datetime(data_frame['pickup_datetime'])
    data_frame['dropoff_datetime'] = pd.to_datetime(data_frame['dropoff_datetime'])
    data_frame['trip_duration'] = (data_frame['dropoff_datetime'] - data_frame['pickup_datetime']).dt.total_seconds()
    return data_frame

# Apply function 1
example_data = calculate_trip_duration(example_data)
print("Output of calculate_trip_duration:")
print(example_data[['pickup_datetime', 'dropoff_datetime', 'trip_duration']])

# Function 2: add_time_features
def add_time_features(data_frame):
    data_frame['pickup_datetime'] = pd.to_datetime(data_frame['pickup_datetime'])
    data_frame['hour_of_day'] = data_frame['pickup_datetime'].dt.hour
    data_frame['day_of_week'] = data_frame['pickup_datetime'].dt.dayofweek
    return data_frame

# Apply function 2
example_data = add_time_features(example_data)
print("\nOutput of add_time_features:")
print(example_data[['pickup_datetime', 'hour_of_day', 'day_of_week']])

# Function 3: calculate_mean_and_margin_of_error
def calculate_mean_and_margin_of_error(data_frame, confidence_level=0.95):
    grouped_data = data_frame.groupby(['PULocationID', 'DOLocationID', 'day_of_week', 'hour_of_day'])['trip_duration']
    mean_duration = grouped_data.mean()
    standard_error = grouped_data.std() / np.sqrt(grouped_data.count())
    margin_of_error = t.ppf((1 + confidence_level) / 2, grouped_data.count() - 1) * standard_error

    # Replace NaN values in margin_of_error with an empty string
    margin_of_error = margin_of_error.apply(lambda x: '' if np.isnan(x) else x)

    predictions = pd.DataFrame({
        'mean_duration': mean_duration,
        'margin_of_error': margin_of_error
    })
    return predictions

# Apply function 3
predictions = calculate_mean_and_margin_of_error(example_data)
print("\nOutput of calculate_mean_and_margin_of_error:")
print(predictions)
```

```python
from scipy.stats import norm

lower_bound = 0.10
upper_bound = 1.0
mean = 0.15
std_dev = 0.028

result = norm.cdf(upper_bound, loc=mean, scale=std_dev) - norm.cdf(lower_bound, loc=mean, scale=std_dev)
print(result)
```
